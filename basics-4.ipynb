## Question 1:

_Assuming that we have some email addresses in the "username@companyname.com" format, please write program
to print the company name of a given email address._

_Both user names and company names are composed of letters only._

__Input Format:__

The first line of the input contains an email address.

__Output Format:__

Print the company name in single line.

__Example;__

_Input: john@google.com_

_Output: google_

```python
import re
def dispCompany(st):
    x = re.match(r'(\w+)@(\w+)\.', st)
    
    if x is not None and x.group(2).isalpha():
        print("The associated company is: {0}".format(x.group(2)))
    else:
        print("Pattern is not a valid email address or invalid domain")
    

def _test():
    iList = ['abcxyz@testcomp.co.uk', 
             'john@google.com',
             'gkrish@syna.co.ps',
             'negativedomain.com',
             'negative@test',
             'testdomainnu@world234.com',
             'invalid12234']
    [dispCompany(s) for s in iList]
    
def main():
    email = input("Provide the email address: ")
    print()
    dispCompany(email)
    
_test()
main()
```

## Question 2:

_Write a program that accepts a comma-separated sequence of words as input and prints the words in a comma
separated sequence after sorting them alphabetically._

__Input Format:__

The first line of input contains words separated by the comma.

__Output Format:__

Print the sorted words separated by the comma.

___Example:___

_Input: without,hello,bag,world_

_Output: bag,hello,without,world_

```python
def ordWord(sl): 
    return (sorted(sl))

def _test():
    _tList = [
        ["without", "hello", "bag", "world"],
        ["winter", "summer", "fall", "autum", "wind"],
    ]
    
    _oList = [
        ["bag", "hello", "without", "world"],
        ["autum", "fall", "summer", "wind", "winter"],
    ]
    
    for i in range(len(_tList)):
        got = ordWord(_tList[i])
        
        assert (got == _oList[i]), "Mismatch\n\t Got: {0}\n\t Want: {1}".format(got, _oList[i])

_test()       
```

### ___Understanding about Python Sets___

    

Set is an unordered & unindex collection of data, Will be denoted with curly {} brackets

__Supported Keywords and functions of a Set__

"in" keyword: To provide iterators access and validation of the data inside a set

"add" method: To add one item to the set

"update" method: To add multiple items

"len" method: To provide the total elements in the set

"remove" method: To remove a particular item from the set, Raise an error/exception if the item is not present in the set

"discard" method: Discard a particular item from the set, do not raise any error/exception if the item is not available

"pop" method: Similar to a Stack, it removes the last item from the set

"clear" method: This helps to destroy the entire set elements

"del" keyword: This serves the same purpose, where it deletes the set from the current context

Apart from the above, it provides the basic elements of a set with Intersection, Union, Disjoint and Update on various set parameters

Let us focus few of the examples below



### ___Programs for Demonstration___

```python
## Can not hold duplicates and un ordered & un indexed
a = {"abc", "one", "ten", "ninetysix", "abc"}
print("a is: {0}".format(a))

b = {"ninetyone", "to", "ninetysix", "ten", "world"}
print("b is: {0}".format(b))

b.add("five")

a.update(["two", "times", "two", "is", "four"])
print("a got updated: {0}".format(a))

for x in b:
    print(x, end=", ")

print("{0} in b".format("five","five" in b))
print("{0} in a".format("seven","seven" in b))

assert (a.remove("is") is not KeyError), "Raising an error???"

a.discard("never")

# c is aUb
c = a.union(b)
print("c is union of a & b: {0}".format(c))

# d is aNb (intersect)
d = a.intersection(b)
print("d is intersection of a & b: {0}".format(d))

print("d is subset of a: {0}".format(d.issubset(a)))
print("c is superset of b: {0}".format(c.issuperset(b)))

print("c is disjoint of d: {0}".format(c.isdisjoint(d)))

e = a.symmetric_difference(b)
print("e is a symmetric difference between a and b: {0}".format(e))

print("c and d difference update: {0}".format(c.difference_update(d)))

print("c and d intersection update: {0}".format(c.intersection_update(d)))

print("pop from d: {0}".format(d.pop()))

f = b.copy()
print("f is copy of b", end=": ")
for x in f:
    print(x, end=", ")

c.clear()
print("\nc is cleared, so empty set: {0}".format(c))

del a, b, c, d, e, f
```

## Question 4:

_Given a list of n-1 numbers ranging from 1 to n, your task is to find the missing number. There are no duplicates._

__Input Format:__

The first line contains n-1 numbers with each number separated by a space.

__Output Format:__

Print the missing number

___Example:___

_Input:1 2 4 6 3 7 8_

_Output:5_

__Explanation:__

In the above list of numbers 5 is missing and hence 5 is the input


```python
'''
    Assumptions:
        1. Always the value 1 and n-1 will be present in the list
        2. No more than one missing value
'''

from functools import reduce
def misSeries(lList):
    if min(lList) != 1:
        return 0
    
    fSum = lambda x: (x * (x + 1) // 2)
    mis = (fSum(max(lList)) - reduce(lambda x, y: x + y, lList))
    if mis > max(lList):
        mis = 0   
    if mis in lList:
        return 0
    
    return mis

def _test():
    tList = [
        [1, 2, 5, 4, 6, 3, 8],
        [1, 3, 5, 2, 6, 7, 8, 9],
        [1, 2, 4, 5, 6, 7],
        [2, 4, 3, 5, 7, 8],
        [1, 4, 3, 2, 6, 7, 8, 9],
        [1, 2, 3, 4, 5, 6, 7],
        [1, 4, 5, 6, 7, 8, 9],
        [1, 2, 3, 5, 6, 7, 9],
    ]
    
    oList = [7, 4, 3, 0, 5, 0, 0, 0]

    for i in range(0, len(tList)):
        x = misSeries(tList[i])
        if x != oList[i]:
            print(x)
        assert ( x == oList[i] ), "Mismatch:\n\t Got:\t {0}\n\t Want:\t {1}".format(x, oList[i])
        
_test()
```

## Question 5:

_With a given list L, write a program to print this list L after removing all duplicate values with original order reserved._

___Example:___

_If the input list is:_ 12 24 35 24 88 120 155 88 120 155

_Then the output should be:_ 12 24 35 88 120 155

__Explanation:__

Third, seventh and ninth element of the list L has been removed because it was already present.

_Input Format:_ In one line take the elements of the list L with each element separated by a space.

_Output Format:_ Print the elements of the modified list in one line with each element separated by a space.

___Example:___

_Input:_ 12 24 35 24

_Output:_ 12 24 35

```python
## Assuming the Max Size of the bitmap as constant
import time

def dupRemoveDict(lList):
    return (list(dict.fromkeys(lList)))

def dupRemove(lList):
    # Taking the bitmap to a upper bound of 4G elements, will take about 33 bytes in python
    BMAP = ( 1 << 32)
    out = []
    
    for i in lList:
        if (BMAP & (1 << i)) == 0:
            BMAP |= (1 << i)
            out.append(i)
    return out            

def _test():
    tList = [
        [12, 24, 35, 24],
        [12, 24, 35, 24, 88, 120, 155, 88, 120, 155],
        [1, 5, 4, 6, 5, 3, 9, 6, 2],
    ]
    
    oList = [
        [12, 24, 35],
        [12, 24, 35, 88, 120, 155],
        [1, 5, 4, 6, 3, 9, 2],
    ]
    
    st = time.time()
    itr = 100000
    for _ in range(itr):
        for i in range(0, len(tList)):
            got = dupRemove(tList[i])
            assert(oList[i] == got), "Mismatch found:\n\t Got:\t {0}, \n\t Want:\t {1}".format(got, oList[i])
    print("DupRemove ran for {0} seconds for {1} iterations".format((time.time() - st), itr))
    
    st = time.time()
    for _ in range(itr):
        for i in range(0, len(tList)):
            got = dupRemoveDict(tList[i])
            assert(oList[i] == got), "Mismatch found:\n\t Got:\t {0}, \n\t Want:\t {1}".format(got, oList[i])
    print("DupRemoveDict ran for {0} seconds for {1} iterations".format((time.time() - st), itr))
        

def main():
    inList = input("Enter the input list: ")
    inList = [int(x) for x in inList.split()]
    print(*("{0} ".format(x) for x in dupRemoveDict(inList)))
    
_test()
main()
```
